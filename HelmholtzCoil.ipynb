```python
import numpy as np
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.gca(projection='3d')

I = 1
a = 2 #coil radius
b = 1  #dist of coil center from xy plane. Coils spaced by 2b then.
mu0 = 1
B0 = -3*mu0*I* a**2 * b / (2*(a**2 + b**2)**(5/2))


X = np.linspace(-a/2, a/2, 5)
Y = np.linspace(-a/2, a/2, 5)
Z = np.linspace(-1.25, 1.25, 5)

x,y,z = np.meshgrid(X, Y, Z)


# Derived for points far from the coil (close to the axis)
# So keep x, y << a
Bx = B0*x
By = B0*y
Bz = B0*-2*z


ax.quiver(x,y,z, Bx, By, Bz)
ax.set_zlabel("z", fontsize=20)
ax.set_ylabel("y", fontsize=20)
ax.set_xlabel("x", fontsize=20)
```

```python
X = np.linspace(-2, 2, 11)
Y = np.linspace(-2, 2, 11)
Z = np.linspace(-2, 2, 11)
```
